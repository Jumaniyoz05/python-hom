```python
# Vazifa: Ob-havo API'siUshbu URL manzilidan foydalaning:
#  https://openweathermap.org/Aniq shahar uchun ob-havo ma'lumotlarini olish uchun requests kutubxonasini ishlating
#  (masalan, sizning tug'ilgan shahringiz: Toshkent) va tegishli ma'lumotlarni (harorat, namlik va boshqalar) chop eting.

import requests

# API kalit (o'z kalitingizni shu yerga yozing)
api_key = 'YOUR_API_KEY'

# Ob-havo API URL
city = 'Tashkent'
url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'

# So‚Äòrov yuboramiz
response = requests.get(url)

# JSON javobni o‚Äòqib olamiz
data = response.json()

# Agar so‚Äòrov muvaffaqiyatli bo‚Äòlsa
if 200<=response.status_code <=299:
    print(f"Shahar: {data['name']}")
    print(f"Harorat: {data['main']['temp']}¬∞C")
    print(f"Namlik: {data['main']['humidity']}%")
    print(f"Bosim: {data['main']['pressure']} hPa")
    print(f"Havo holati: {data['weather'][0]['description']}")
else:
    print("Xatolik yuz berdi:", data.get("message", "Noma'lum xatolik"))
```

```python
#Vazifa: JSONni o'zgartirish Dastur yozing, bu foydalanuvchilarga yangi kitob qo'shishga, 
# mavjud kitob ma'lumotlarini yangilashga va books.json 
# JSON faylida kitoblarni o'chirishga imkon beradi.

import json
import os

# JSON fayl nomi
filename = 'books.json'

# JSON faylni yuklab olish
def load_books():
    if os.path.exists(filename):
        with open(filename, 'r', encoding='utf-8') as file:
            return json.load(file)
    return []

# JSON faylga saqlash
def save_books(books):
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(books, file, ensure_ascii=False, indent=4)

# üìò Yangi kitob qo‚Äòshish
def add_book():
    books = load_books()
    new_id = max([book['id'] for book in books], default=0) + 1
    nomi = input("Kitob nomi: ")
    muallif = input("Muallifi: ")
    yil = int(input("Chop etilgan yili: "))
    
    new_book = {
        "id": new_id,
        "nomi": nomi,
        "muallif": muallif,
        "yil": yil
    }
    
    books.append(new_book)
    save_books(books)
    print("‚úÖ Kitob qo‚Äòshildi!")

# ‚úèÔ∏è Kitobni yangilash
def update_book():
    books = load_books()
    id_to_update = int(input("Yangilamoqchi bo‚Äòlgan kitob IDsi: "))
    
    for book in books:
        if book['id'] == id_to_update:
            book['nomi'] = input("Yangi nomi: ")
            book['muallif'] = input("Yangi muallifi: ")
            book['yil'] = int(input("Yangi yil: "))
            save_books(books)
            print("‚úÖ Kitob yangilandi!")
            return
    
    print("‚ùå Bunday IDli kitob topilmadi.")

# ‚ùå Kitobni o‚Äòchirish
def delete_book():
    books = load_books()
    id_to_delete = int(input("O‚Äòchirmoqchi bo‚Äòlgan kitob IDsi: "))
    
    updated_books = [book for book in books if book['id'] != id_to_delete]
    
    if len(books) == len(updated_books):
        print("‚ùå Bunday IDli kitob topilmadi.")
    else:
        save_books(updated_books)
        print("üóëÔ∏è Kitob o‚Äòchirildi.")

# üìã Menyu
def show_menu():
    while True:
        print("\n--- Kitoblar Boshqaruvi ---")
        print("1. Yangi kitob qo‚Äòshish")
        print("2. Kitobni yangilash")
        print("3. Kitobni o‚Äòchirish")
        print("4. Chiqish")
        tanlov = input("Tanlovingiz (1-4): ")

        if tanlov == '1':
            add_book()
        elif tanlov == '2':
            update_book()
        elif tanlov == '3':
            delete_book()
        elif tanlov == '4':
            print("Dasturdan chiqildi.")
            break
        else:
            print("‚ö†Ô∏è Noto‚Äòg‚Äòri tanlov!")

# Dastur ishga tushadi
show_menu()






```

```python
#  Vazifa: Film Tavsiya Tizimi Ushbu URL manzilidan http://www.omdbapi.com/ film haqida ma'lumot olish uchun foydalaning.
#  Foydalanuvchilardan film janrini so'raydigan va o'sha janrdan tasodifiy film tavsiya qiladigan dastur yarating.

import requests
import random

API_KEY = 'YOUR_API_KEY'  # Bu yerga o'z API kalitingizni yozing

films_by_genre = {
    "Action": ["John Wick", "Mad Max: Fury Road", "Gladiator"],
    "Comedy": ["The Mask", "Superbad", "The Hangover"],
    "Drama": ["Forrest Gump", "The Shawshank Redemption", "The Godfather"],
    "Horror": ["The Conjuring", "It", "A Quiet Place"],
    "Sci-Fi": ["Inception", "The Matrix", "Interstellar"]
}

print("Mavjud janrlar:", ', '.join(films_by_genre.keys()))
genre = input("Qaysi janrdan film tavsiya qilay? ").title()

if genre in films_by_genre:
    film_nomi = random.choice(films_by_genre[genre])
    url = f"http://www.omdbapi.com/?apikey={API_KEY}&t={film_nomi}"
    response = requests.get(url)
    data = response.json()

    if data["Response"] == "True":
        print("\nüé¨ Tavsiya qilinadigan film:")
        print(f"üéû Nomi: {data['Title']}")
        print(f"üìÖ Yili: {data['Year']}")
        print(f"‚≠ê Reyting: {data['imdbRating']}")
        print(f"üé≠ Janr: {data['Genre']}")
        print(f"üìñ Tavsif: {data['Plot']}")
    else:
        print("‚ùå Film topilmadi.")
else:
    print("‚ùå Bunday janr mavjud emas.")







```
