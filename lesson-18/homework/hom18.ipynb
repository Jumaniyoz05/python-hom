```python
# Titanik ma'lumotlar to'plami, CSV formatida saqlangan.  Ma'lumotlar quyidagi ma'lumotlar ustunlaridan iborat.
import pandas as pd
titanic_df = pd.read_csv(r'C:\Users\hp\Downloads\titanic.csv')
titanic_df.head()

# PassengerId: Har bir yo'lovchining ID raqami.
# 20 dan 30 yoshgacha bo'lgan 1-sinfdagi ayol yo'lovchilarni tanlang: 
# 20 dan 30 yoshgacha bo'lgan 1-sinfdagi ayol yo'lovchilarni o'z ichiga olgan DataFrame ni ajrating.
# Select Female Passengers in Class 1 with Ages between 20 and 30: 
# Extract a DataFrame containing female passengers in Class 1 with ages between 20 and 30.
titanic_df[(
    titanic_df['Age'].between(20,30))&
    (titanic_df['Pclass']==1)&
    (titanic_df['Sex']=='femele'
)]

```

```python
# Filter Passengers Who Paid More than $100: Create a DataFrame with passengers who paid a fare greater than $100.
# 100 dollardan ortiq to'lov qilgan yo'lovchilarni filtrlash: 
# 100 dollardan ko'p to'lov qilgan yo'lovchilar bilan DataFrame yarating.

yuzdan_ortiq=titanic_df[titanic_df['Fare']>=100.0]
yuzdan_ortiq
```

```python
# Select Passengers Who Survived and Were Alone: 
# Filter passengers who survived and were traveling alone (no siblings, spouses, parents, or children).
# Yuzaga kelgan va yolg‘iz bo‘lgan yo‘lovchilarni tanlang: 
# Yolg‘iz (uka-uka, turmush o‘rtog‘i, ota-onalari yoki bolalari bo‘lmagan) va omon qolgan yo‘lovchilarni filtrlang.
omon=titanic_df[(titanic_df['Survived']==1)&
           (titanic_df['SibSp']==0)&
           (titanic_df['Parch']==0)]
omon
```

```python
# Filter Passengers Embarked from 'C' and Paid More Than $50:
#  Create a DataFrame with passengers who embarked from 'C' and paid more than $50.
#C' dan yo'l olgan va 50 dollardan ortiq to'lagan yo'lovchilarni filtrlang:
#  'C' dan yo'l olgan va 50 dollardan ortiq to'lagan yo'lovchilar uchun DataFrame yarating.
c_dagilar=titanic_df[(titanic_df['Embarked']=='C')&
           (titanic_df['Fare']>=50.0)
           ]
c_dagilar
```

```python
# Select Passengers with Siblings or Spouses and Parents or Children:
#  Extract passengers who had both siblings or spouses aboard and parents or children aboard.
# Akalar yoki turmush o'rtog'lari va ota-onalari yoki bolalari bo'lgan yo'lovchilarni tanlang: 
# akalari yoki turmush o'rtog'lari va ota-onalari yoki bolalari bo'lgan yo'lovchilarni ajrating.
akasi_bor=titanic_df[(titanic_df['Survived']==1)&
           (titanic_df['SibSp']==1)&
           (titanic_df['Parch']==1)]
akasi_bor

```

```python
# Filter Passengers Aged 15 or Younger Who Didn't Survive:
#  Create a DataFrame with passengers aged 15 or younger who did not survive.

# 15 yosh yoki undan kichik bo'lgan va omon qolmagan yo'lovchilarni filtrlash: 
# 15 yosh yoki undan kichik bo'lgan va omon qolmagan yo'lovchilar bilan DataFrame yarating.
olgan_bolaalr=titanic_df[(titanic_df['Survived']==0)&
           (titanic_df['Age']<=15)]
olgan_bolaalr
```

```python
# Select Passengers with Cabins and Fare Greater Than $200:
#  Extract passengers with known cabin numbers and a fare greater than $200.

# Kabinalari va tariflari $200 dan baland bo'lgan yo'lovchilarni tanlang: 
# Kabin raqamlari ma'lum bo'lgan va tariflari $200 dan yuqori bo'lgan yo'lovchilarni oling.

titanic_df[(titanic_df['Cabin']!='nan')&
           (titanic_df['Fare']>=200)]
```
