```python
# Kelgusi tug'ilgan kunigacha qoldi kunlar:
#  Birinchi mashqga o'xshash, lekin bu safar foydalanuvchining kelgusi tug'ilgan
#  kunigacha qoldi kunlar sonini hisoblang va chop eting.
from datetime import datetime, date, timedelta

# Foydalanuvchidan tug‘ilgan sanani olish (faqat oy va kun)
birth_month = int(input("Tug‘ilgan oyingizni kiriting (1-12): "))
birth_day = int(input("Tug‘ilgan kuningizni kiriting (1-31): "))

# Bugungi sana
today = date.today()
this_year = today.year

# Tug‘ilgan kun - bu yil
next_birthday = date(this_year, birth_month, birth_day)

# Agar tug‘ilgan kun bu yil allaqachon o‘tgan bo‘lsa, keyingi yilga olamiz
if next_birthday < today:
    next_birthday = date(this_year + 1, birth_month, birth_day)

# Farqni hisoblash
days_left = (next_birthday - today).days

print(f"Kelgusi tug‘ilgan kuningizgacha {days_left} kun qoldi.")


```

```python
# Uchrashuvni rejalashtiruvchi: Foydalanuvchidan joriy sana va vaqtni, 
# shuningdek, uchrashuv davomiyligini soat va daqiqalarda kiritishni so'rang.
#  Uchrashuv qachon tugashini hisoblang va chop eting.

from datetime import datetime, timedelta

# Foydalanuvchidan joriy sana va vaqtni olish
sana_va_vaqt = input("Joriy sana va vaqtni kiriting (YYYY-MM-DD HH:MM): ")

# Stringni datetime obyektga aylantiramiz
joriy_vaqt = datetime.strptime(sana_va_vaqt, "%Y-%m-%d %H:%M")

# Uchrashuv davomiyligini olish
soat = int(input("Uchrashuv davomiyligi - soatlar: "))
daqiqa = int(input("Uchrashuv davomiyligi - daqiqalar: "))

# Uchrashuv davomiyligi obyektini yaratish
davomiylik = timedelta(hours=soat, minutes=daqiqa)

# Tugash vaqtini hisoblash
tugash_vaqti = joriy_vaqt + davomiylik

# Natijani chiqarish
print(f"Uchrashuv {tugash_vaqti.strftime('%Y-%m-%d %H:%M')} da tugaydi.")
```

```python
# Vaqt zonasini o'zgartiruvchi: 
# Foydalanuvchidan sana va vaqtni, shuningdek, hozirgi vaqt zonasini kiritishni talab qiluvchi dastur 
# yarating, so'ngra kiritilgan vaqtni foydalanuvchi tanlagan boshqa vaqt zonasida o'zgartirib, chop eting.

#  Vaqt zonasi konvertori: foydalanuvchiga joriy vaqt zonasi bilan birga
#  sana va vaqtni kiritishga imkon beradigan dastur yarating, so'ngra sana
#  va vaqtni o'zi tanlagan boshqa vaqt zonasiga o'zgartiring va chop eting.


from datetime import datetime  # Sana va vaqt bilan ishlash uchun
import pytz  # Vaqt zonalari bilan ishlash uchun

# 1. Foydalanuvchidan joriy vaqt zonasini olish
current_tz_name = input("Joriy vaqt zonangizni kiriting (masalan, Asia/Tashkent): ")

# 2. Sana va vaqtni matndan datetime obyektiga aylantirish
date_str = input("Sana va vaqtni kiriting (YYYY-MM-DD HH:MM): ")
date_format = "%Y-%m-%d %H:%M"
naive_datetime = datetime.strptime(date_str, date_format)  # vaqt zonasiz datetime

# 3. Foydalanuvchidan yangi vaqt zonasini olish
new_tz_name = input("Yangi vaqt zonangizni kiriting (masalan, America/New_York): ")

# 4. Vaqt zonalari obyektlarini yaratamiz
current_tz = pytz.timezone(current_tz_name)
new_tz = pytz.timezone(new_tz_name)

# 5. Sana va vaqtni joriy vaqt zonasiga moslashtirish
current_tz_datetime = current_tz.localize(naive_datetime)

# 6. Sana va vaqtni yangi vaqt zonasiga o'tkazish
new_tz_datetime = current_tz_datetime.astimezone(new_tz)

# 7. Natijani chiqarish
print(f"Joriy vaqt zonangizdagi sana va vaqt: {current_tz_datetime.strftime(date_format)}")
print(f"Tanlangan yangi vaqt zonasidagi sana va vaqt: {new_tz_datetime.strftime(date_format)}")

pytz.common_timezones
```

```python
import re

# Email uchun regex shabloni
email_regex = r'^[\w\.-]+@[\w\.-]+\.\w{2,}$'

# Foydalanuvchidan email olish
email = input("Email manzilini kiriting: ")

# Regex orqali tekshirish
if re.match(email_regex, email):
    print("✅ Email manzili to'g'ri formatda.")
else:
    print("❌ Noto‘g‘ri email formati. Iltimos, qayta kiriting.")
```

```python
# Telefon raqami formatlovchi: Telefon raqamini kiriting va uni standart formatga mos ravishda
# formatlovchi dastur yarating. Masalan, "1234567890" ni "(123) 456-7890" ga aylantiring.

import re

def format_phone_number(phone):
    # Faqat raqamlar (0–9) ajratiladi
    digits = re.sub(r'\D', '', phone)

    # 10 ta raqam bo'lishi kerak
    if len(digits) != 10:
        return "❌ Noto‘g‘ri telefon raqami. 10 ta raqam kiriting."

    # regex orqali formatlash
    formatted = re.sub(r"(\d{3})(\d{3})(\d{4})", r"(\1) \2-\3", digits)
    return f"✅ Formatlangan raqam: {formatted}"

# Foydalanuvchi kiritishi
user_input = input("Telefon raqamini kiriting: ")
print(format_phone_number(user_input))


```

```python
#So'z Izlagich: Berilgan matnda ma'lum bir so'zning barcha takrorlanishlarini topadigan dastur tuzing. 
# Foydalanuvchidan so'z kiritishni so'rang, so'ngra namuna matnda 
# o'sha so'zning barcha takrorlanishlarini qidiring va chop eting.

import re

# 1. Namuna matn
matn = "Bu yerda kod yozish mashq qilinmoqda. Kod yozish ko'nikmasi muhim."

# 2. Foydalanuvchidan so'z kiritishni so'rash
soz = input("Qanday so'zni qidiramiz? ")

# 3. So'zni matndan topish (case-insensitive qidiruv)
# \b - so'z chegarasini bildiradi, shunda "kod" so'zi "kodlash" ichidan topilmaydi
mosliklar = re.findall(rf'\b{re.escape(soz)}\b', matn, re.IGNORECASE)

# 4. Natijani chiqarish
if mosliklar:
    print(f"Topildi: {len(mosliklar)} marta")
    for i, topilgan in enumerate(mosliklar, 1):
        print(f"{i}) {topilgan}")
else:
    print("So'z topilmadi.")


```

```python
import re

# 1. Foydalanuvchidan matn kiritishni so'rash
matn = input("Matnni kiriting: ")

# 2. Sana formatlarini aniqlash uchun regex (kun-oy-yil, misol: 24.08.2025 yoki 24-08-2025)
# \d{2} - 2 xonali raqam (kun va oy)
# \d{4} - 4 xonali raqam (yil)
sana_pattern = r'\b\d{2}[./-]\d{2}[./-]\d{4}\b'

# 3. Matndan sanalarni qidirish
sanalar = re.findall(sana_pattern, matn)

# 4. Natijani chiqarish
if sanalar:
    print(f"Topilgan sanalar ({len(sanalar)} ta):")
    for i, sana in enumerate(sanalar, 1):
        print(f"{i}) {sana}")
else:
    print("Hech qanday sana topilmadi.")
```
