```python
# Bir Odam klasi yaratish uchun Python dasturini yozing. 
# Ism, mamlakat va tug'ilgan kun kabi atributlarni qo'shing.
# Odamning yoshini aniqlash uchun bir usulni amalga oshiring.

from datetime import datetime

class Odam:
    def __init__(self, ism, mamlakat, tugilgan_kun):
        self.ism = ism
        self.mamlakat = mamlakat
        # tugilgan_kun 'YYYY-MM-DD' formatidagi satr deb hisoblaymiz
        self.tugilgan_kun = datetime.strptime(tugilgan_kun, "%Y-%m-%d").date()

    def yoshini_hisobla(self):
        bugun = datetime.today().date()
        yosh = bugun.year - self.tugilgan_kun.year

        # Tug'ilgan kun hali kelmagan bo'lsa yoshni 1 ga kamaytiramiz
        if (bugun.month, bugun.day) < (self.tugilgan_kun.month, self.tugilgan_kun.day):
            yosh -= 1
        return yosh

# Misol uchun obyekt yaratish va yoshini chiqarish
odam1 = Odam("Jumaniyoz", "O'zbekiston", "2005-05-20")
print(f"{odam1.ism}ning yoshi: {odam1.yoshini_hisobla()}")




```

```python
#  Ikkilik qidirish daraxtini ifodalovchi sinf yaratish uchun Python dasturini yozing.
#  Ikkilik daraxtda elementlarni qo'shish va qidirish usullarini kiritishni o'z ichiga oling.

# Daraxt tuguni (node) sinfi
class Node:
    def __init__(self, value):
        self.value = value      # Tugun qiymati
        self.left = None        # Chap farzand
        self.right = None       # O‘ng farzand

# Ikkilik qidiruv daraxti (BST) sinfi
class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Element qo‘shish funksiyasi
    def insert(self, value):
        new_node = Node(value)
        if self.root is None:
            self.root = new_node
            return
        
        current = self.root
        while True:
            if value < current.value:
                if current.left is None:
                    current.left = new_node
                    return
                current = current.left
            else:
                if current.right is None:
                    current.right = new_node
                    return
                current = current.right

    # Element qidirish funksiyasi
    def search(self, value):
        current = self.root
        while current is not None:
            if value == current.value:
                return True
            elif value < current.value:
                current = current.left
            else:
                current = current.right
        return False

# Foydalanish misoli:
bst = BinarySearchTree()
bst.insert(50)
bst.insert(30)
bst.insert(70)
bst.insert(20)
bst.insert(40)
bst.insert(60)
bst.insert(80)

print(bst.search(60))  # True
print(bst.search(25))  # False
```

```python
#  7 Write a Python program to create a class representing a linked list data structure.
#  Include methods for displaying linked list data, inserting, and deleting nodes.

# Python dasturini yozing, bu bog'langan ro'yxat ma'lumotlar strukturasi hisoblangan sinfni yaratadi.
# Bog'langan ro'yxat ma'lumotlarini ko'rsatish, tugunlarni qo'shish va o'chirish uchun metodlarni qo'shing.

class Node:
    """Bog'langan ro'yxatdagi har bir tugunni ifodalovchi sinf"""
    def __init__(self, data):
        self.data = data  # Tugundagi ma'lumot
        self.next = None  # Keyingi tugunga havola


class LinkedList:
    """Bog'langan ro'yxat sinfi"""
    def __init__(self):
        self.head = None  # Ro'yxat boshini belgilovchi ko'rsatkich

    def add(self, data):
        """Ro'yxat oxiriga yangi tugun qo'shish"""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def delete(self, key):
        """Berilgan qiymatga ega birinchi tugunni o'chirish"""
        current = self.head
        prev = None

        while current and current.data != key:
            prev = current
            current = current.next

        if current is None:
            print(f"{key} ro'yxatda topilmadi.")
            return

        if prev is None:
            self.head = current.next  # Bosh tugun o'chirilyapti
        else:
            prev.next = current.next

    def display(self):
        """Ro'yxatdagi barcha elementlarni chiqarish"""
        current = self.head
        if current is None:
            print("Ro'yxat bo'sh.")
            return

        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


# Namuna ishlatish
if __name__ == "__main__":
    lst = LinkedList()
    lst.add(10)
    lst.add(20)
    lst.add(30)
    print("Ro'yxat:")
    lst.display()

    print("\n20 ni o'chiramiz:")
    lst.delete(20)
    lst.display()

    print("\n40 ni o'chirishga urinamiz:")
    lst.delete(40)
    lst.display()


```
